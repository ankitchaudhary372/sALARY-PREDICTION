```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model = LinearRegression()
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv("rent.csv")
```

```python
data.head(3)
```

```python
scaler = MinMaxScaler()
```

```python
data["Size_Scale"] = scaler.fit_transform(data[["Size"]])
```

```python
data
```

```python
encode = LabelEncoder()
```

```python
data["Point_of_contact_Lable"]= encode.fit_transform(data[["Point of Contact"]])
data["Area Type_Lable"]= encode.fit_transform(data[["Area Type"]])
data["City_Lable"]= encode.fit_transform(data[["City"]])
data["Furnishing Status_Lable"]= encode.fit_transform(data[["Furnishing Status"]])
data["Tenant Preferred_Lable"]= encode.fit_transform(data[["Tenant Preferred"]])
```

```python
data.head(4)
```

```python
X = data[["Size_Scale"]]  
```

```python
X = data[["Size_Scale","Point_of_contact_Lable","Area Type_Lable","City_Lable","Furnishing Status_Lable","Tenant Preferred_Lable"]]  
```

```python
X
```

```python
Y = data[["Rent"]]
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y , test_size =0.3 , random_state =41)
```

model = LinearRegression()
model.fit(X_train,Y_train)

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),      # optional, scales numeric features
    ("lr", LinearRegression())
])
```

```python
pipe.fit(X_train, Y_train) 
```

```python
y_pred = pipe.predict(X_test)  
```

```python
print("R2:", r2_score(Y_test, y_pred))
print("MSE:", mean_squared_error(Y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Predictions for Train and Test
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

plt.figure(figsize=(12,5))

# Train Plot
plt.subplot(1,2,1)
plt.scatter(y_train, y_train_pred)
plt.xlabel("Actual Rent (Train)")
plt.ylabel("Predicted Rent")
plt.title("Train: Actual vs Predicted")
plt.plot([y_train.min(), y_train.max()],
         [y_train.min(), y_train.max()],
         linestyle='--')

# Test Plot
plt.subplot(1,2,2)
plt.scatter(y_test, y_test_pred)
plt.xlabel("Actual Rent (Test)")
plt.ylabel("Predicted Rent")
plt.title("Test: Actual vs Predicted")
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         linestyle='--')

plt.tight_layout()
plt.show()
```
